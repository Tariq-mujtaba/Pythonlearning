```python
# You are tasked with creating a Python program that serves as an interactive tool for a friend who enjoys exploring numbers. The program should begin by prompting the user to enter their name and then ask them for three of their favorite numbers.
import math


friendname:str = input("Enter your name: ")

num1: int = int(input("Enter your first fav number: "))
num2: int = int(input("Enter your second fav number: "))
num3: int = int(input("Enter your third fav number: "))

#  After gathering this information, the program should greet the user with a personalized message that includes their name. 
print(f"Hello, {friendname}! Let's explore your favorite numbers:") 
 
# The three numbers provided by the user should be stored in a list. 
nums:list = [num1, num2, num3]

# The program should then check if any of the numbers are even or odd, and store this information in a separate list of tuples, where each tuple contains the number and a string indicating whether it is "even" or "odd".
evenorodd:list=[]

for num in nums: 
    if((num%2)==0):
        evenorodd.append((num, "even"))
    else:
        evenorodd.append((num, "odd"))


print(evenorodd)

#  Following this, the program should use a for loop to iterate over the list of numbers, and for each number, it should create a tuple containing the number and its square. 
squares: list = []

for number in nums: 
    squares.append((number, number*number))

print(squares)

# These tuples should be printed in a creative and engaging format.
print(f"The number {evenorodd[0][0]} is {evenorodd[0][1]}.")
print(f"The number {evenorodd[1][0]} is {evenorodd[1][1]}.")
print(f"The number {evenorodd[2][0]} is {evenorodd[2][1]}.")


print(f"the square of {squares[0][0]} is {squares[0][1]}")
print(f"The sqaure of {squares[1][0]} is {squares[1][1]}")
print(f"The sqaure of {squares[2][0]} is {squares[2][1]}")


#  Additionally, the program should calculate the sum of the three numbers and print the result, accompanied by an encouraging message. 
sumofnums:int = num1+num2+num3
print(f"Amazing! The sum of your favorite numbers is: {sumofnums}.")



# Finally, the program should determine if the sum is a prime number and notify the user with an appropriate message. The goal is to make the tool both enjoyable and informative, allowing the user to explore their favorite numbers in a fun and interactive way, while also introducing some interesting logical checks.
#if number is divible by only 1 and itself it's prime //

if sumofnums > 1:  # 1 is not a prime number
    is_prime = True
    for i in range(2, int(math.sqrt(sumofnums)) + 1):
        if sumofnums % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"wow, {sumofnums} is a prime number.")
    else:
        print(f"{sumofnums} is not a prime number.")
else:
    print(f"{sumofnums} is not a prime number.")
```
